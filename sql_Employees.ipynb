### **Problem Statement**
* You are a data analyst at a tech company. The HR department needs insights about employee data. You have access to an employees table with the following structure:

### Employees Table

| Column        | Data Type | Description                          |
|---------------|-----------|--------------------------------------|
| employee_id   | INTEGER   | Unique identifier (Primary Key)      |
| name          | TEXT      | Employee name                        |
| department    | TEXT      | Department name                      |
| salary        | INTEGER   | Annual salary                        |
| hire_date     | DATE      | Date of joining                      |

The table contains 100 employee records across departments: Engineering, Sales, Marketing, Finance, HR, and Operations.

```python
import sqlite3
conn = sqlite3.connect("Employee.db")
cursor = conn.cursor()
print("Employee Database created Successfully")
```

```python
cursor.execute ("""
            create table employees (employee_id int primary key ,
                                    emp_name text ,
                                    department text ,
                                    salary int ,
                                    hire_date date);
                  """)

conn.commit()
```

* I created a new SQLite database called Employee.db.
* Inside it, we created a table called employees with columns for ID, name, department, salary, and joining date.
* employee_id is unique for each employee.

```python
# The table contains 100 employee records across departments: Engineering, Sales, Marketing, Finance, HR, and Operations.
# insert employees record
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()
departments = ["Engineering", "Sales", "Marketing", "Finance", "HR", "Operations"]
for i in range(1,101):
    name = fake.name()
    dept = random.choice(departments)
    salary = random.randint(40000,120000)
    hire_date = fake.date_between(start_date='-5y', end_date='today').strftime("%Y-%m-%d")  # convert to string

    cursor.execute("""
    INSERT INTO employees VALUES (?, ?, ?, ?, ?)
    """,(i,name,dept,salary,hire_date))

conn.commit()
```

* I installed Faker, a Python tool to generate realistic fake data.

* Generated 100 employees with random names, departments, salaries, and hire dates.

* Each employee is added to the database with a unique ID.

# Task 1: Basic Querying

```python
# Write SQL queries to answer the following:
# 1. Select all columns for employees in the Marketing department
marketing_dept = pd.read_sql_query("select * from employees where department = 'Marketing' ",conn)
marketing_dept
```

* Shows all employee details who work in Marketing.
* Helps HR quickly find everyone in that department.

```python
# 2. Display only name, department, and salary for employees earning more than $90,000
high_salary = pd.read_sql_query("select emp_name ,department,salary from employees where salary > 90000 order by salary DESC ", conn)
high_salary
```

* Shows only name, department, and salary for employees earning above 90000.
* Sorted from highest to lowest salary to easily see top earners.

```python
# 3. Find employees in either Sales OR Finance departments
sal_fin_dept = pd.read_sql_query("select * from employees where department = 'Sales'  OR department = 'Finance'",conn)
sal_fin_dept
```

* Shows all details for employees in Sales or Finance.
* Using IN makes it simpler than multiple OR conditions.

# Task 2: Sorting and Limiting

```python
# Write SQL queries to find:
# 1. The top 5 highest-paid employees (sort by salary descending, limit to 5)
top_emp = pd.read_sql_query("select * from employees order by salary DESC limit 5",conn)
top_emp
```

* Finds the top 5 employees with the highest salaries.

* Useful for spotting top performers or budgeting for raises.

```python
# 2. The 3 most recently hired employees in Engineering department (sort by hire_date descending)
most_engg_dept = pd.read_sql_query("select * from employees where department = 'Engineering' order by hire_date DESC limit 3",conn)
most_engg_dept
```

* Shows the latest 3 employees who joined Engineering.

* Helps HR track recent hiring trends.

```python
# 3.The oldest employee (by hire_date) who earns more than $70,000 and is NOT in HR department
old_emp = pd.read_sql_query("select * from employees where salary > 70000 and department NOT in ('Engineering') order by hire_date  ",conn)
old_emp
```

* Finds the longest-serving employee earning more than 70000 but not in HR.
* Useful for reward programs or mentoring plans.

# Task 3: Calculated Fields and Aliases

```python
# 1. Calculate monthly salary for all employees:
#  Display name, salary, and salary/12 aliased as monthly_pay
month_salary = pd.read_sql_query("select emp_name as Name ,salary as Annual_pay , salary/12 as Monthly_pay from employees",conn)
month_salary
```

* Calculates monthly salary by dividing annual salary by 12.
*  Shows name, annual pay, and monthly pay for all employees.

```python
#  Only show employees earning more than $60,000
earn_more_salary = pd.read_sql_query("select emp_name as Name ,salary as Annual_pay , salary/12 as Monthly_pay from employees where salary >60000",conn)
earn_more_salary
```

* Same as above but only for employees earning more than 60000.
* Helps HR focus on higher-earning employees for financial planning.

```python
# 2. Create a report for the Finance department showing:
# name aliased as employee_name # salary aliased as annual_salary
# salary/12 aliased as monthly_salary # Sort by salary descending
salary_Report= pd.read_sql_query("select emp_name as employee_Name ,salary as Annual_Salary , salary/12 as Monthly_Salary from  employees WHERE department = 'Finance' order by salary DESC",conn)
salary_Report
```

* Reports Finance employees with annual & monthly salaries, sorted high → low.
* Useful for budgeting, salary review, or departmental analysis.
